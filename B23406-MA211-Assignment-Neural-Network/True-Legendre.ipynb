**`This code is used to cross validate the legendre's solution as it is the exact solution of legendre which I have implemented using TF`**

```python
# Using matlab got this code and for getting exact legendre graph

import numpy as np
import tensorflow as tf

def legendre_polynomial(n, x):
    # Ensure x is a NumPy array if it's a TensorFlow tensor
    if isinstance(x, tf.Tensor):
        x = x.numpy()

    # Initialize the first two Legendre polynomials
    p0 = np.ones_like(x)  # P_0(x) = 1
    p1 = x                # P_1(x) = x

    # Handle the case for n = 0 and n = 1 explicitly
    if n == 0:
        return p0
    elif n == 1:
        return p1

    # Calculate higher-order Legendre polynomials using the recurrence relation
    for k in range(2, n + 1):
        p2 = ((2 * k - 1) * x * p1 - (k - 1) * p0) / k
        p0, p1 = p1, p2  # Update for the next iteration

    return p1
```

```python
x_values = np.linspace(-1, 1, 1000)
n = 6
legendre_vals = legendre_polynomial(n, x_values)

# Optionally plot the results using Matplotlib
import matplotlib.pyplot as plt

plt.plot(x_values, legendre_vals)
plt.title(f'Legendre Polynomial P_{n}(x)')
plt.xlabel('x')
plt.ylabel(f'P_{n}(x)')
plt.grid()
plt.show()
```
